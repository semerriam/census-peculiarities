## New Jersey

##### 2020 Redistricting Data for New Jersey at the Block Level (Adjusted) 

From DataHub ReadMe file: These datasets were retrieved from the New Jersey Department of State at https://nj.gov/state/dos-reports-2020-census-report.shtml The RDH retrieved the data from the source listed and did not modify any of the data. The data for the Block level is available by county. There is one XLSX file for each county, and one sheet per table (P1, P2, P3, P4). All XLSX files are included in this zipped folder.

Importing in Atlantic County

```python
df_nj2 = pd.read_excel("..//data//source//RDH-data//nj_pl2020_b_official_xlsx//atlantic-county-blocks_R2.xlsx")
df_nj2.head(18)
```

```python
df_nj2 = df_nj2.rename(columns={'P1. POPULATION BY RACE AND HISPANIC OR LATINO ORIGIN (INCLUDING RE-ALLOCATED INCARCERATED POPULATION)': 'State'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 1': 'County'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 2': 'Municipality'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 3': 'Tract'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 4': 'Group'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 5': 'Block'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 6': 'VTD'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 7': 'VTDI'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 8': 'County Name'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 9': 'Municipality Name'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 10': 'Areaname'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 11': 'Population'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 12': 'Total'})
df_nj2 = df_nj2.rename(columns={'Unnamed: 22': '18plus'})
```

```python
df_nj2 = df_nj2.drop(df_nj2.index[range(17)])
```

```python
df_nj2_subset = df_nj2[['State', 'County', 'Municipality', 'Tract', 'Group', 'Block', 'VTD', 'VTDI', 'County Name', 'Municipality Name', 'Areaname', 'Population', 'Total', '18plus']]
df_nj2_subset.head(5)
```
